# Problem 7

Identify the three blocks in this code:

spam = 0

if spam == 10:

print('eggs')

if spam > 5:

print('bacon')

else:

print('ham')

print('spam')

print('spam')


```python
spam = 10
if spam == 10:
    print('eggs')
if spam > 5:
    print('bacon')
else:
    print('ham')
print('spam')
print('spam')
```

### Solution



```python
spam = 0
if spam == 10:
    print('eggs')        # indent increased, block 1
    if spam > 5:         # still block 1
        print('bacon')   # still block 1, indent increased, block 2 inside block 1
    else:                # still block 1, indent decreased, block 2 ended in line above
        print('ham')     # still block 1, indent increased, block 3 inside block 1
    print('spam')        # still block 1, indent decreased, block 3 ended in line above
print('spam')            # indent decreased, block 1 ended in line above
```

# Problem 10
How can you tell the difference between break and continue?

### Solution

The break statement takes care of terminating the loop in which it is used. If the break statement is used inside nested loops, the current loop is terminated, and the flow will continue with the code followed that comes after the loop.

The continue statement skips the code that comes after it, and the control is passed back to the start for the next iteration.

# Problem 11 

In a for loop, what is the difference between range(10), range(0, 10), and range(0, 10, 1)?



### Solution

There is no difference in these three.

the syntex of range is range(start, end, step). 

start is optional, default value is zero.

step value is also optional, Default is 1. 

```python
for i in range(10):
    print(i)
```

```python
for i in range(0,10):
    print(i)
```

```python
for i in range(0,10,1):
    print(i)
```
